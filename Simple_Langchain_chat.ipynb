```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
```

```python
from langchain_groq import ChatGroq

# Initialize Groq LLM
llm = ChatGroq(
    model= "llama-3.3-70b-versatile", #"llama-3.1-70b-versatile",  # or "mixtral-8x7b", "gemma2-9b-it", etc.
    temperature=0.7,
    max_tokens=1024,
   
)

print(llm)
```

```python
result = llm.invoke("Tell me top 5 tourist places of Mangalore")
```

```python
result.content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","you are AI engineer, answer the question with accurate information"),
        ("user","{input}")
    ]
)
```

```python
chain = prompt|llm
output = chain.invoke({"input": "What are top 5 tourist places of mangalore"})
```

```python
output.content
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser() #Post-processing step that ensures your LLM response is returned as plain text instead of a structured object. 

chain=prompt|llm|output_parser
```

```python

output = chain.invoke({"input": "What are top 5 tourist places of mangalore. provide in bullet points"})
output = output.replace("\\n", "\n")

print(output)
```
